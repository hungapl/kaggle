```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as stats

%matplotlib inline

x = np.arange(10)
y = x+10

data = pd.DataFrame({'x':x,'y':y})
plt.figure(figsize=(16, 6))
sns.scatterplot(x='x', y='y', data=data).set_title('Example scatterplot')
```

# Distribution Plot

```python
data = np.random.lognormal(size=100)
plt.show()
sns.distplot(data, kde=False, fit=stats.norm)
```

```python
sns.distplot(data, kde=False, fit=stats.lognorm)
plt.show()
```
